```python
Annual_Salary_cols = ['Annual_Salary']  

for col in Annual_Salary_cols:
    df[col] = (
        df[col]
        .astype(str)  
        .str.replace(r'[^\d.]', '', regex=True)  
        .replace('', pd.NA)  
    )
    df[col] = pd.to_numeric(df[col], errors='coerce')  
```

```python
Timestamp_cols = ['Timestamp']
for col in Timestamp_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
    mean_date = df[col].dropna().mean()  
    df[col].fillna(mean_date, inplace=True)
```

Outliers & treatments


Annual_Salary

```python
no_mis_df = df.dropna(axis=0)
no_mis_df.duplicated().sum()
```

```python
no_mis_df.drop_duplicates(keep ="first", inplace=True)
```

```python
sns.boxplot(data=no_mis_df,x="Annual_Salary")
```

```python
Q1 = no_mis_df["Annual_Salary"].quantile(0.25)
Q3 = no_mis_df["Annual_Salary"].quantile(0.75)

IQR = Q3-Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Create arrays of Boolean values indicating the outlier rows

upper_array = np.where(no_mis_df["Annual_Salary"] >= upper) [0]
lower_array = np.where(no_mis_df["Annual_Salary"] <= lower) [0]
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

no_mis_df["Annual_Salary"] = np.where(
    no_mis_df["Annual_Salary"] > upper,
    upper,
    np.where(
        no_mis_df["Annual_Salary"] < lower,
        lower,
        no_mis_df["Annual_Salary"])
)
```

```python
sns.boxplot(data=no_mis_df,x="Annual_Salary")
```

```python
sns.boxplot(data=no_mis_df,x="Compensation")
```
